```python
import yfinance as yf

tesla = yf.Ticker("TSLA")
tesla_stock_data = tesla.history(period="5y")

# Reset the index
tesla_data = tesla_stock_data.reset_index()

# Save the dataframe to a CSV file
tesla_data.to_csv('tesla_data.csv', index=False)

# Display the first five rows
print(tesla_data.head())
```

```python
import requests
from bs4 import BeautifulSoup
import pandas as pd

url = "https://www.macrotrends.net/stocks/charts/TSLA/tesla/revenue"
html = requests.get(url).text
soup = BeautifulSoup(html, 'html.parser')
table = soup.find_all('table')[1]
rows = table.find_all('tr')[1:]

tesla_revenue_data = []
for row in rows:
    columns = row.find_all('td')
    date = columns[0].text.strip()
    revenue = columns[1].text.strip()
    tesla_revenue_data.append((date, revenue))

tesla_revenue = pd.DataFrame(tesla_revenue_data, columns=['Date', 'Revenue'])

# Save the dataframe to a CSV file
tesla_revenue.to_csv('tesla_revenue.csv', index=False)

# Display the last five rows
print(tesla_revenue.tail())
```

```python
import plotly.graph_objects as go

def make_graph(stock_data, title):
    fig = go.Figure(go.Scatter(x=stock_data['Date'], y=stock_data['Close'], mode='lines'))
    fig.update_layout(title=title, xaxis_title='Date', yaxis_title='Close Price')
    fig.show()

tesla_data = pd.read_csv('tesla_data.csv')
make_graph(tesla_data, 'Tesla Stock Price')
```

```python
gme_data = pd.read_csv('gme_data.csv')
make_graph(gme_data, 'GameStop Stock Price')
```
